```python
import numpy as np
import pandas as pd

# Data for the 8 planets in our Solar System
# X = Distance from Sun in AU (Astronomical Units)
# y = Orbital Period in Earth Years
planets = {
    "Name": ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune"],
    "Distance_AU": [0.387, 0.723, 1.000, 1.524, 5.203, 9.537, 19.191, 30.069],
    "Period_Years": [0.241, 0.615, 1.000, 1.881, 11.862, 29.457, 84.011, 164.79]
}

df = pd.DataFrame(planets)

# PySR expects X to be a 2D array (shaped like a table)
X = df[['Distance_AU']].values
y = df['Period_Years'].values

print("Dataset ready!")
print(df)
```

```python
from pysr import PySRRegressor

# This is the 'Brain' configuration
model = PySRRegressor(
    niterations=100,             # How many 'generations' of equations to evolve
    binary_operators=["+", "-", "*", "/", "^"], # Basic math + Powers (like x^2)
    unary_operators=["square", "cube"],         # Common physics patterns
    model_selection="best",      # Pick the most accurate but simple one
    loss="loss(prediction, target) = (prediction - target)^2", # Standard error

    # This prevents 'Math Spaghetti' (equations that are too long to read)
    maxsize=20,

    # We tell it to prioritize simple integer powers
    complexity_of_operators={"^": 3, "square": 2, "cube": 2}
)

# Start the 'Evolution'
model.fit(X, y)
```

```python
import matplotlib.pyplot as plt

# Get the best equation from the model
best_eqn = model.get_best()
predictions = model.predict(X)

plt.figure(figsize=(10, 6))

# Plot the real data points
plt.scatter(df['Distance_AU'], df['Period_Years'], color='red', label='Actual Planets', zorder=5)

# Plot the AI's discovered law
# We create a smooth line of distances to show the curve
x_smooth = np.linspace(0, 31, 500).reshape(-1, 1)
y_smooth = model.predict(x_smooth)
plt.plot(x_smooth, y_smooth, label=f'AI Discovered Law: $T \propto R^{{{best_eqn.equation[5:10]}}}$', color='blue', linewidth=2)

plt.xlabel('Distance from Sun (AU)')
plt.ylabel('Orbital Period (Years)')
plt.title('Kepler\'s Third Law: AI Rediscovery from Solar System Data')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import pandas as pd

def lennard_jones(r, epsilon=1.0, sigma=1.0):
    # The famous 12-6 potential law
    return 4 * epsilon * ((sigma/r)**12 - (sigma/r)**6)

# Generate 1000 random "molecule" configurations
np.random.seed(42)
num_samples = 1000

# Let's assume two critical distances (r1 and r2) define our system energy
r1 = np.random.uniform(0.9, 2.5, num_samples)
r2 = np.random.uniform(0.9, 2.5, num_samples)

# The "Hidden Law" the AI needs to find:
# Energy = LJ(r1) + LJ(r2) + a small interaction term
y_energy = lennard_jones(r1) + lennard_jones(r2) + 0.5 * (r1 * r2)

# Add "Measurement Noise" (makes it arXiv-tier realistic)
y_noisy = y_energy + np.random.normal(0, 0.05, num_samples)

X = np.stack([r1, r2], axis=1)

print("Synthetic Research Dataset Generated!")
print(f"X shape (Distances): {X.shape}")
print(f"y shape (Energy): {y_noisy.shape}")
```

```python
import sympy
from pysr import PySRRegressor

model = PySRRegressor(
    niterations=100,
    binary_operators=["+", "-", "*", "/", "^"],
    unary_operators=["inv(x) = 1/x"],
    # We tell SymPy that "inv" should be written as 1/x
    extra_sympy_mappings={"inv": lambda x: 1/x},

    # Updated parameter name to avoid FutureWarnings
    elementwise_loss="loss(prediction, target) = (prediction - target)^2",

    constraints={'^': (-1, 12)},
    model_selection="best"
)

# Start the discovery
model.fit(X, y_noisy)
```

```python
import matplotlib.pyplot as plt

# 1. Get the list of equations and their metrics
equations = model.equations_

# 2. Plot Complexity vs Log(Loss)
plt.figure(figsize=(10, 6))
plt.plot(equations['complexity'], np.log10(equations['loss']), 'o-', markersize=8, linewidth=2, color='#2c3e50')

# Highlight the "Best" equation
best_idx = 17
plt.scatter(equations.iloc[best_idx]['complexity'], np.log10(equations.iloc[best_idx]['loss']),
            color='red', s=200, label='Selected Physical Law', zorder=5)

plt.xlabel('Equation Complexity (Number of Operators)', fontsize=12)
plt.ylabel('Log10(Mean Squared Error)', fontsize=12)
plt.title('The Pareto Front: Accuracy vs. Simplicity in Potential Discovery', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Add an annotation for the "Elbow"
plt.annotate('Optimal Trade-off',
             xy=(equations.iloc[11]['complexity'], np.log10(equations.iloc[11]['loss'])),
             xytext=(18, -0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

```python
noise_levels = [0.01, 0.05, 0.1, 0.2] # 1%, 5%, 10%, 20% noise
results = []

for noise in noise_levels:
    print(f"Testing Noise Level: {noise*100}%...")

    # Generate data with specific noise
    y_test = y_energy + np.random.normal(0, noise, num_samples)

    # Run a quick search
    temp_model = PySRRegressor(
        niterations=30, # Faster iterations for the sensitivity test
        binary_operators=["+", "-", "*", "/", "^"],
        unary_operators=["inv(x) = 1/x"],
        extra_sympy_mappings={"inv": lambda x: 1/x},
        elementwise_loss="loss(prediction, target) = (prediction - target)^2",
        verbosity=0
    )
    temp_model.fit(X, y_test)

    # Log the best loss found
    best_loss = temp_model.get_best().loss
    results.append({"Noise_Level": noise, "Best_Loss": best_loss})

# Display as a scientific table
results_df = pd.DataFrame(results)
print("\n--- NOISE SENSITIVITY TABLE ---")
print(results_df)
```

```python
# Stop the previous cell manually (the [ ] button in Colab)
# Then run this optimized version

model = PySRRegressor(
    niterations=50,             # Fewer, more focused generations
    populations=20,             # More 'parallel' guesses
    binary_operators=["+", "-", "*", "/", "^"],
    unary_operators=["inv(x) = 1/x"],
    extra_sympy_mappings={"inv": lambda x: 1/x},

    # We use a 'Parabolic' complexity cost to discourage extreme exponents
    complexity_of_operators={"^": 4, "inv": 2},

    constraints={'^': (12, 1)},
    elementwise_loss="loss(prediction, target) = (prediction - target)^2",

    # This prevents the search from getting stuck on one bad idea
    tournament_selection_n=12,
)

model.fit(X, y_noisy)
```

```python
import seaborn as sns

# 1. Generate predictions from your best equation
y_pred = model.predict(X)
residuals = y_noisy - y_pred

plt.figure(figsize=(12, 5))

# Subplot 1: Parity Plot (The Gold Standard)
plt.subplot(1, 2, 1)
plt.scatter(y_noisy, y_pred, alpha=0.5, color='teal')
plt.plot([y_noisy.min(), y_noisy.max()], [y_noisy.min(), y_noisy.max()], 'r--', lw=2)
plt.xlabel('Actual Energy (Ground Truth + Noise)')
plt.ylabel('AI Predicted Energy')
plt.title('Parity Plot: Ground Truth vs. AI')

# Subplot 2: Residual Distribution
plt.subplot(1, 2, 2)
sns.histplot(residuals, kde=True, color='purple')
plt.axvline(0, color='black', linestyle='--')
plt.xlabel('Error (Actual - Predicted)')
plt.title('Residual Analysis (Error Distribution)')

plt.tight_layout()
plt.show()
```
